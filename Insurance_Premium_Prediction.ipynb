# Linear Regression

```python
# Scikit Learn module
from sklearn.linear_model import LinearRegression


lin_reg = LinearRegression()
lin_reg.fit(X_train,y_train)

y_pred_LR = lin_reg.predict(X_test)

#Evaluvation: MSE
from sklearn.metrics import mean_squared_error
J_mse_sk_LR = mean_squared_error(y_pred_LR, y_test)

# R_square
R_square_sk_LR = lin_reg.score(X_test,y_test)

print('The Mean Square Error(MSE) or J(theta) is: ',J_mse_sk_LR)
print('R square obtain for scikit learn library is :',R_square_sk_LR)
```

# Lasso Linear RegressionÂ¶


```python
# Scikit Learn module
from sklearn.linear_model import Lasso


lin_reg_lasso = Lasso()
lin_reg_lasso.fit(X_train,y_train)

y_pred_LR_lasso = lin_reg_lasso.predict(X_test)

#Evaluvation: MSE
from sklearn.metrics import mean_squared_error
J_mse_sk_lasso = mean_squared_error(y_pred_LR_lasso, y_test)

# R_square
R_square_sk_lasso = lin_reg_lasso.score(X_test,y_test)

print('The Mean Square Error(MSE) or J(theta) is: ',J_mse_sk_lasso)
print('R square obtain for scikit learn library is :',R_square_sk_lasso)
```

```python
J_mse_sk_lasso
```

# Ridge Linear Regression

```python
# Scikit Learn module
from sklearn.linear_model import Ridge


lin_reg_ridge = Ridge()
lin_reg_ridge.fit(X_train,y_train)

y_pred_LR_ridge = lin_reg_ridge.predict(X_test)

#Evaluvation: MSE
from sklearn.metrics import mean_squared_error
J_mse_sk_LR_ridge = mean_squared_error(y_pred_LR_ridge, y_test)

# R_square
R_square_sk_LR_ridge = lin_reg_ridge.score(X_test,y_test)

print('The Mean Square Error(MSE) or J(theta) is: ',J_mse_sk_LR_ridge)
print('R square obtain for scikit learn library is :',R_square_sk_LR_ridge)
```

# Support Vector Machine


```python
# Support Vector Machine's 
from sklearn.svm import SVR

SVM_R = SVR(kernel='rbf')
SVM_R.fit(X_train, y_train)

y_pred_svm = SVM_R.predict(X_test)

#Evaluvation: MSE
from sklearn.metrics import mean_squared_error
J_mse_sk_svm = mean_squared_error(y_pred_svm, y_test)

# R_square
R_square_sk_svm = SVM_R.score(X_test,y_test)

print('The Mean Square Error(MSE) or J(theta) is: ',J_mse_sk_svm)
print('R square obtain for scikit learn library is :',R_square_sk_svm)
```

```python
MSE = [J_mse_sk_LR,J_mse_sk_LR_ridge,J_mse_sk_lasso,J_mse_sk_svm,RM]
R_square = [R_square_sk_LR, R_square_sk_LR_ridge,R_square_sk_lasso,R_square_sk_svm,Random_forset]
models = ['Linear Regression', 'Linear Regression (Ridge)' , 'Linear Regression (Lasso)', 'Support Vector Machine','Random forest']
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,
                     scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```
